# QUESTION 2: Use the following specs to extract features from the textual data:
## A. Use the \english" stopwords of the CountVectorizer
## B. Exclude terms that are numbers (e.g. \123", \-45", \6.7" etc.)
## C. Perform lemmatization with nltk.wordnet.WordNetLemmatizer and pos tag
## D. Use min df=3
## Report the shape of the TF-IDF matrices of the train and test subsets respectively.

```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS, CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer


nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

comp_categories = [ 'comp.graphics', 'comp.os.ms-windows.misc', 'comp.sys.ibm.pc.hardware', 
                   'comp.sys.mac.hardware']
rec_categories = ['rec.autos', 'rec.motorcycles', 'rec.sport.baseball', 'rec.sport.hockey']
train_dataset = fetch_20newsgroups(subset = 'train', categories = comp_categories + rec_categories, 
                                  shuffle = True, random_state = 42,)
test_dataset = fetch_20newsgroups(subset = 'test', categories = comp_categories + rec_categories, 
                                 shuffle = True, random_state = 42,)
counts = []

wnl = nltk.wordnet.WordNetLemmatizer()
stop_words = ENGLISH_STOP_WORDS
analyzer = CountVectorizer().build_analyzer()

def penn2morphy(penntag):
    """ Converts Penn Treebank tags to WordNet. """
    morphy_tag = {'NN':'n', 'JJ':'a',
                  'VB':'v', 'RB':'r'}
    try:
        return morphy_tag[penntag[:2]]
    except:
        return 'n'

def lemmatize_sent(list_word):
    # Text input is string, returns array of lowercased strings(words).
    return [wnl.lemmatize(word.lower(), pos=penn2morphy(tag)) 
            for word, tag in pos_tag(list_word)]

def lemmatize_rmv_nums(doc):
    #To remove terms that are numbers(e.g. "123", "-45", "6.7" etc.)
    return (word for word in lemmatize_sent(analyzer(doc)) if word not in stop_words and not word.isdigit())

analyzer = CountVectorizer().build_analyzer()
vectorizer = CountVectorizer(min_df = 3, analyzer = lemmatize_rmv_nums, stop_words = 'english')

X_train = vectorizer.fit_transform(train_dataset.data)
X_test = vectorizer.transform(test_dataset.data)

tfidf_T = TfidfTransformer()

X_train_tfidf = tfidf_T.fit_transform(X_train)# TF-IDF matrix of the train subsets
X_test_tfidf = tfidf_T.transform(X_test)# TF-IDF matrix of the test subsets

print('Shpae of train subsets: ', X_train.shape)
print('Shape of TF-IDF matrix of train subsets: ', X_train_tfidf.shape)
print('Shape of test subsets: ', X_test.shape)
print('Shape of TF-IDF matrix of test subsets: ', X_test_tfidf.shape)
```

# Question 3:
## Reduce the dimensionality of the data using the methods above

## A. Apply LSI to the TF-IDF matrix corresponding to the 8 categories with k = 50; so each document is mapped to a 50-dimensional vector.
## B. Also reduce dimnsionality through NMF (k = 50) and compare with LSI:

### Which one is larger, the $ \mid\mid X-WH\mid\mid_F^2 $ in NMF or the $\mid\mid X-U_k\sum_kV_k^T\mid\mid_F^2$ in LSI? Why is the case?

## LSA/LSI

```python
#Apply LSI to TF-IDF matrices of train subsets
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components = 50, random_state = 42)
X_train_reduced = svd.fit_transform(X_train_tfidf)
X_test_reduced = svd.transform(X_test_tfidf)
V = svd.components_

print(X_train_reduced.shape)
print(X_test_reduced.shape)
print(V.shape)
```

## NMF: Non-negative matrix factorization
\begin{array}{cc}
\min_\limits{\mathbf{W}, \mathbf{H}} & \lVert\mathbf{X} - \mathbf{W}\mathbf{H}\rVert_F^2 \\
s.t. & \mathbf{W} \ge 0 \\
& \mathbf{H} \ge 0
\end{array}

$\mathbf{X} \in \mathbb{R}^{m\times n}$,

$\mathbf{W} \in \mathbb{R}^{m\times k}$, $\mathbf{H} \in \mathbb{R}^{k\times n}$, $k < \mathrm{rank}(\mathbf{X})$

```python
#Apply NMF to TF-IDF matrices of train subsets
from sklearn.decomposition import NMF

model = NMF(n_components = 50, init = 'random', random_state = 42)

W_train = model.fit_transform(X_train_tfidf)
W_test = model.fit_transform(X_test_tfidf)
print(W_train.shape)
print(W_test.shape)

H = model.components_
print(H.shape)
```

### Compare $ \mid\mid X-WH\mid\mid_F^2 $ and $\mid\mid X-U_k\sum_kV_k^T\mid\mid_F^2$

```python
LSI_train_e = np.sum(np.array(X_train_tfidf - X_train_reduced.dot(V))**2)
print('LSI training error: ', LSI_train_e)
NMF_train_e = np.sum(np.array(X_train_tfidf - W_train.dot(H))**2)
print('NMF training error: ', NMF_train_e)

LSI_test_e = np.sum(np.array(X_test_tfidf - X_test_reduced.dot(V))**2)
print('LSI testing error: ', LSI_test_e)

NMF_test_e = np.sum(np.array(X_test_tfidf - W_test.dot(H))**2)
print('NMF testing error: ', NMF_test_e)

if LSI_train_e > NMF_train_e:
    print('LSI traingin error is larger.')
elif NMF_train_e > LSI_train_e:
    print('NMF training error is larger.')
else:
    print('LSI and NMF are equal.')
    
if LSI_test_e > NMF_test_e:
    print('LSI testing error is larger.')
elif NMF_test_e > LSI_test_e:
    print('NMF testing error is larger.')
else:
    print('LSI and NMF are equal.')
```

```python
#import string
label_train = train_dataset.target > 3
label_train = label_train.astype(int)
label_test = test_dataset.target > 3
label_test = label_test.astype(int)

#label_train = get_labels(train_dataset)
#label_test = get_labels(test_dataset)

# check
print('-'*20,'train_dataset','-'*20)
print(len(label_train))
print(np.unique(label_train))
print('-'*20,'test_dataset','-'*20)
print(len(label_test))
print(np.unique(label_test))
```

```python
#Train SVM with two different gamma(gamma_1 = 1000(hard margin), gamma_2 = 0.0001(soft margin))
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
import matplotlib.pyplot as plt

svm_hard = LinearSVC(C = 1000, random_state = 42).fit(X_train_reduced, label_train)

def plot_roc(fpr, tpr):
    fig, ax = plt.subplots()

    roc_auc = metrics.auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label= 'area under curve = %0.4f' % roc_auc)

    ax.grid(color='0.7', linestyle='--', linewidth=1)

    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate',fontsize=15)
    ax.set_ylabel('True Positive Rate',fontsize=15)

    ax.legend(loc="lower right")

    for label in ax.get_xticklabels()+ax.get_yticklabels():
        label.set_fontsize(15)
        
score_hard = svm_hard.decision_function(X_test_reduced)
print(len(score_hard))
fpr_hard, tpr_hard, _ = metrics.roc_curve(label_test, score_hard)
plot_roc(fpr_hard, tpr_hard)

predicted_hard = svm_hard.predict(X_test_reduced)
```

```python
svm_soft = LinearSVC(C = 0.0001, random_state = 42).fit(X_train_reduced, label_train)

def plot_roc(fpr, tpr):
    fig, ax = plt.subplots()

    roc_auc = metrics.auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label= 'area under curve = %0.4f' % roc_auc)

    ax.grid(color='0.7', linestyle='--', linewidth=1)

    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate',fontsize=15)
    ax.set_ylabel('True Positive Rate',fontsize=15)

    ax.legend(loc="lower right")

    for label in ax.get_xticklabels()+ax.get_yticklabels():
        label.set_fontsize(15)

score_soft = svm_soft.decision_function(X_test_reduced)
print(len(score_soft))
fpr_soft, tpr_soft, threshold_soft = metrics.roc_curve(label_test, score_soft)
plot_roc(fpr_soft, tpr_soft)

predicted_soft = svm_soft.predict(X_test_reduced)
```

```python
fig, ax = plt.subplots()

line_hard = ax.plot(fpr_hard, tpr_hard, linewidth=2)
line_soft = ax.plot(fpr_soft, tpr_soft, linewidth=2)

ax.grid(color='0.7', linestyle='--', linewidth=1)

ax.set_xlim([-0.1, 1.1])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate',fontsize=15)
ax.set_ylabel('True Positive Rate',fontsize=15)

ax.legend(('Hard SVM','Soft SVM'),loc="lower right")
plt.title('ROC curve of SVM with C=1000(hard) and C=0.0001(soft)')

for label in ax.get_xticklabels()+ax.get_yticklabels():
    label.set_fontsize(15)
```

```python
import itertools

def plot_confusion_matrix(cm, title, cmap = plt.cm.Blues):
    ctgrs = ['Class I \n Computers','Class II \n Recreation']
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(ctgrs))
    plt.xticks(tick_marks, ctgrs, rotation=45)
    plt.yticks(tick_marks, ctgrs)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    
confusion_hard = confusion_matrix(label_test, predicted_hard)
confusion_soft = confusion_matrix(label_test, predicted_soft)
```

```python
# Plot non-normalized confusion matrices
plot_confusion_matrix(confusion_soft, title='Soft confusion matrix')
plot_confusion_matrix(confusion_hard, title='Hard confusion matrix')

#Print 'Accuracy', 'Precision', 'Recall', 'F1' of hard margin
print('-'*10,'SVM_Hard','-'*10)
print('Accuracy: ', metrics.accuracy_score(label_test, predicted_hard))
print('Precision:', metrics.precision_score(label_test, predicted_hard))
print('Recall:   ', metrics.recall_score(label_test, predicted_hard))
print('F1:       ', metrics.f1_score(label_test, predicted_hard))

#Print 'Accuracy', 'Precision', 'Recall', 'F1' of soft margin
print('-'*10,'SVM_Soft','-'*10)
print('Accuracy: ', metrics.accuracy_score(label_test, predicted_soft))
print('Precision:', metrics.precision_score(label_test, predicted_soft))
print('Recall:   ', metrics.recall_score(label_test, predicted_soft))
print('F1:       ', metrics.f1_score(label_test, predicted_soft))
```

```python
#Finding optimal gamma
from sklearn.model_selection import cross_val_score
scores_k = []
for k in range(-3,4):
    #Finding the best value of gamma in the range 10^k, -3 <= k <= 3
    clf_svm = LinearSVC(C=10**k).fit(X_train_reduced, label_train)
    scores = cross_val_score(clf_svm, X_train_reduced, label_train, cv=5, scoring='accuracy')
    scores_k.append(np.mean(scores))
k_opt = 10**(np.argmax(scores_k)-3)
print(k_opt)
```

```python
#Plotting roc curves of optimal gamma
svm_opt = LinearSVC(C = k_opt, random_state = 42).fit(X_train_reduced, label_train)
predict_opt = svm_opt.predict(X_test_reduced)
score_opt = svm_opt.decision_function(X_test_reduced)
fpr_opt, tpr_opt, thresholds_opt = roc_curve(label_test, score_opt)

def plot_roc(fpr, tpr):
    fig, ax = plt.subplots()

    roc_auc = metrics.auc(fpr, tpr)

    ax.plot(fpr, tpr, lw=2, label= 'area under curve = %0.4f' % roc_auc)

    ax.grid(color='0.7', linestyle='--', linewidth=1)

    ax.set_xlim([-0.1, 1.1])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate',fontsize=15)
    ax.set_ylabel('True Positive Rate',fontsize=15)

    ax.legend(loc="lower right")

    for label in ax.get_xticklabels()+ax.get_yticklabels():
        label.set_fontsize(15)
        
plot_roc(fpr_opt, tpr_opt)

#Ploting confusion matrix of optimal gamma
confusion_opt = confusion_matrix(label_test, predict_opt)
plot_confusion_matrix(confusion_opt, title = 'Optimal SVM confusion matrix')

print('-'*10,'SVM_opt','-'*10)
print('Accuracy: ', metrics.accuracy_score(label_test, predict_opt))
print('Precision:', metrics.precision_score(label_test, predict_opt))
print('Recall:   ', metrics.recall_score(label_test, predict_opt))
print('F1:       ', metrics.f1_score(label_test, predict_opt))
```

# QUESTION 7: Grid search of parameters:
## A. Construct a Pipeline that performs feature extraction, dimensionality reduction and classification;
## B. Do grid search with 5-fold cross-validation to compare the following (use test accuracy as the score to compare):

## What is the best combination?

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS, CountVectorizer
from sklearn import linear_model
from sklearn.externals.joblib import Memory
from tempfile import mkdtemp
from shutil import rmtree

pipeline = Pipeline([
    ('vect', CountVectorizer(min_df = 3,
                             analyzer = lemmatize_rmv_nums, 
                             stop_words = 'english')),
    ('tfidf', TfidfTransformer()),
    ('reduce_dim', TruncatedSVD(n_components = 50, random_state = 42)),
    ('clf', GaussianNB())]   
)

param_grid = [
    {
        'vect_analyzer': [lemmatize_rmv_nums, 'word'],
        'vect_min_df': [3, 5],
        'reduce_dim': [TruncatedSVD(n_components = 50, random_state = 42),
                       NMF(n_components=50, init='random', random_state=42)],
        'clf': [LinearSVC(C = 10, random_state = 42),
                GaussianNB()]
    }
]
grid = GridSearchCV(pipeline, cv = 5, n_jobs = -1, param_grid = param_grid, verbose = 5, scoring = 'accuracy')
#print(train_dataset.data.shape)
#print(label_train.shape)
```
